```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_hdf('Ol_TSG_2001-2012.pandas.h5')
df.head()
```

```python
df.tail()
```

```python
lon_fixed = df.lon.copy()
lon_fixed[lon_fixed>0] = -lon_fixed[lon_fixed>0]
df['lon'] = lon_fixed
plt.plot(df.lon, df.lat)
```

```python
df.plot.hexbin(x='lon', y='lat', C='temp', reduce_C_function=np.std, sharex=False)
plt.title('Oleander SST Standard Deviation')
```
